Income            | Tax on this income 
------------------|----------------------

### Algo

    2. Calculate the tax payable


Processing for tax payable calculation:
 - If income is less than 18200, then no tax applied
 - if income is between 18201 and 45000 the (income - 18200)*0.19
 - If income is between 45001 and 120000 then (income - 45000)*0.325 + 5092
 - If income is between 120001 and 180000 then (income -120000 )*0.37 + 29647
 - If income is greater than (income - 180000 )*0.45 + 51667

# Task 2: Step 5 - Convert the pseudocode to Python Code

```python
#promt user for input then conver the input into int data type.

income = 40500
if(income >= 18201 and  income <= 45000):
    taxableAmount = income - 18201
    tax = taxableAmount*0.19
elif(income >= 45001 and  income <= 120000):
    taxableAmount = income - 45000
    tax = (taxableAmount*0.325) + 5092
elif(income >= 120001 and  income <= 180000):
    taxableAmount = income - 120000
    tax = (taxableAmount*0.37) + 29647
elif(income >= 180001):
    taxableAmount = income - 180000
    tax = (taxableAmount*0.45) + 51667
else:
    tax=0
print("No Tax ") if (tax == 0) else print(f'The tax payable is {tax}')
```
