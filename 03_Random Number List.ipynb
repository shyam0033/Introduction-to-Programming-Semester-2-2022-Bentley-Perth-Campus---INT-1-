### Randomly generate numbers and store in two lists: odd and even.
### For each list find the minimum, maximum and average

```python
odd =  []
even= []
lowerLimit = 1
upperLimiit = 500
totalNumber = 25
i= 1

while i <= totalNumber:
    randomNumber =  random.randint(lowerLimit,upperLimiit)
    if (randomNumber%2 == 0):
        even.append(randomNumber)
    else:
        odd.append(randomNumber)
    i=i+1
oddMinNumber  = min(odd) 
evenMinNumber = min(even)  
oddMaxNumber  = max(odd) 
evenMaxNumber = max(even) 
oddAverageNumber  = sum(odd) / len(odd)
evenAverageNumber = sum(even) / len(even)

print(f'Odd List: Avergae = {oddAverageNumber} Minimum Number = {oddMinNumber} Maximum Number = {oddMaxNumber}')
print(f'even List: Avergae = {evenAverageNumber} Minimum Number = {evenMinNumber} Maximum Number = {evenMaxNumber}')


```
