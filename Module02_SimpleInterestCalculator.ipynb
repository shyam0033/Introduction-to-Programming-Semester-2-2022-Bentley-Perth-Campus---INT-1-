# Problem/Requirement:

Bank ABC wants to calculate their client's loan interest using the simple interest method as explained at https://www.bankrate.com/loans/personal-loans/how-to-calculate-loan-interest/

Write a program for bank ABC to compute the loan interest on their client's borrowed amount based on the simple interest method. 






# 1. State Problem Clearly

Compute the loan interest on the borrwed amount . Then display the calculated gross pay

# 2. Describe Inputs and Outputs

```
                     +-----------+
 Loan Amount    ---> |           |
 Interest Rate  ---> | Algorithm | ---> Interest
 Time           ---> |           |
                     +-----------+
```

# 3. Work the problem by hand - worked example

Let's assume 

> Loan Amount = 20000
> Interest Rate = 5% 
> Time = 5       


Manualy, by hand we compute the interest 
Principal loan amount x Interest rate x Time (aka Number of years in term) = Interest


> 20000 * 0.05 * 5 = 5000 

# 4. Develop an Algorithm/Pseudocode

Here is our pseduocode:

```
Get the input variable
Calculate the Interest
Display the Interest
```



# 5. Convert Pseudocode to Python Code

````python
# Get Loan Amount 
LoanAmount = 20000
# Get Interest Rate
InterestRate = 5
# Get No of Years
Years = 5
# Interest = LoanAmount*InterestRate*Time 
Interest = LoanAmount*(InterestRate/100)*Years
# Display The Interest Amount is: gross pay
print(f'Total Interest Amount = ${Interest}')
````

# 6. Test Solution with a variety of data

````python
LoanAmount = float(input("Enter Loan Amount")) 
InterestRate = float(input("Enter Interest Rate Percentagae")) 
Years = float(input("Enter Years")) 
Interest = LoanAmount*(InterestRate/100)*Years
print(f'Total Interest Amount = ${Interest:.2f}')
````
